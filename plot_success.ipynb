```python
import matplotlib.pyplot as plt

# Data from the table
models = ["llama3.3 70b versatile", "gemma1", "gemma2", "mixtral-7b", "GPT-o1 preview", "our method"]
data = {
    "llama3.3 70b versatile": [1, 6, 6, 6, 6, 6, 6, 1, 6, 6, 6, 6, 6, 6],
    "gemma1": [6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6],
    "gemma2": [6, 6, 6, 6, 1, 6, 6, 1, 6, 3, 6, 6, 6, 6],
    "mixtral-7b": [6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6],
    "GPT-o1 preview": [1, 2, 6, 1, 6, 1, 1, 1, 1, 1, 6, 6, 1, 6],
    "our method": [1, 0, 1, 0, 1, 0, 1, 0, 0, 0, 2, 2, 0, 0],
}

# Calculate success rates
success_rates = {}
for model, attempts in data.items():
    successes = sum(1 for attempt in attempts if attempt < 6)
    success_rate = (successes / len(attempts)) * 100
    success_rates[model] = success_rate

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(success_rates.keys(), success_rates.values())

# Customization
plt.title("Levels Success", fontsize=16)
plt.xlabel("Models", fontsize=14)
plt.ylabel("Success Rate (%)", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Display values on bars
for index, value in enumerate(success_rates.values()):
    plt.text(index, value + 2, f"{value:.1f}%", ha='center', fontsize=10)

# Save plot as PDF
plt.tight_layout()
plt.savefig("model_success_rate.pdf")
plt.show()
```
